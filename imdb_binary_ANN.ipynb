**Text Transformation**

```python
def oh_representation(sequences, dim=10000):
  results=np.zeros((len(sequences),dim))
  for i,sequence in enumerate(sequences):
    results[i,sequence]=1
  return results
```

```python
oh_train_data=oh_representation(train_data)
```

```python
oh_train_data.shape
```

```python
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
```

```python
#architechture

model=Sequential()
model.add(Dense(units=16,activation='relu', input_shape=(10000,)))
model.add(Dense(units=16,activation='relu'))
model.add(Dense(units=1,activation='sigmoid',))
```

```python
model.summary()
```

```python
model.compile(optimizer='rmsprop',
              loss="binary_crossentropy",
              metrics=["accuracy"])
```

```python
X_train=oh_representation(train_data)
X_test=oh_representation(test_data)
```

```python
X_val=X_train[:10000]
X_partial_train=X_train[10000:]
y_val=train_labels[:10000]
y_partial_train=train_labels[10000:]
```

```python
history = model.fit(X_partial_train,y_partial_train,
                    validation_data= (X_val,y_val),
                    batch_size=512,
                    epochs=10)
```

```python
history_dict= history.history
```

```python
history_dict.keys()
```

```python
history_dict
```

```python
acc= history_dict['accuracy']
val_acc= history_dict['val_accuracy']
loss= history_dict['loss']
val_loss= history_dict['val_loss']
```

```python
epoch= range(1,len(acc)+1)
plt.plot( epoch, loss, "bo", label = "Training loss")
plt.plot(epoch, val_loss, "--", label = "validation loss")
plt.title("Train loss vs Validation loss")
plt.legend()
plt.show()
```

```python
test_results= model.evaluate(X_test, test_labels)
```
