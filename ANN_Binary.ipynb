```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
```

```python
df=pd.read_csv("/content/income.csv")
df
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
training_data=df.iloc[: 40000, :]
validation_data=df.iloc[40000:45000, : ]
testing_data=df.iloc[45000: ,:]
```

```python
training_data.shape
```

```python
validation_data.shape
```

```python
testing_data.shape
```

```python
training_data_inde=training_data.iloc[: , : -1]
training_data_dep=training_data.iloc[ : , -1]
```

```python
validation_data_inde=validation_data.iloc[: , :-1]
validation_data_dep=validation_data.iloc[ : , -1]
```

```python
testing_data_inde=testing_data.iloc[: , :-1]
testing_data_dep=testing_data.iloc[: ,-1]
```

**Design the Binary ANN Architecture**

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras.activations import relu, softmax
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
model= Sequential()
#input layer ti hiden layer 1
model.add(Dense(units=10,
                kernel_initializer='he_uniform',
                activation="relu" ,
                input_dim=training_data_inde.shape[1]))
#hiden layer 1 to hiden layer 2
model.add(Dense(units=4, kernel_initializer="he_uniform", activation="relu"))
#hiden layer 2 to 3
model.add(Dense(units=2, kernel_initializer='he_uniform', activation='relu'))
#hiden layer 3 to output layer
model.add(Dense(units=1, kernel_initializer='glorot_uniform', activation='sigmoid'))

```

```python
model.compile(optimizer="adam", loss='binary_crossentropy', metrics=['Accuracy'])
```

```python
40000/1000
```

```python
model.summary()
```

```python
#training process
model.fit(training_data_inde,
          training_data_dep,
          batch_size=1000,
          epochs=50,
          validation_data=(validation_data_inde,validation_data_dep))
```

```python
model.weights
```

```python
if model.predict(np.array([[19,124906,10,0,1719,25]]))[0][0]> 0.5:
  print(1)
else:
  print(0)
```

```python
results=model.predict(testing_data_inde)
results
```

```python
c=[]
for i[0] in results:
  if i>0.5:
    c.append(1)
  else:
    c.append(0)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
confusion_matrix(testing_data_dep,c)
```

```python
accuracy_score(testing_data_dep,c)
```

```python
print(classification_report(testing_data_dep, c))
```
